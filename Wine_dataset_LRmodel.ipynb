**Select The features**)

```python
final_cols=['flavanoids','malic_acid','color_intensity','alcohol','proline','class_0']
final_df=df[final_cols]
final_df

```

```python
X=final_df.drop("class_0",axis=1)
y=final_df['class_0']

from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
scaled_df=pd.DataFrame()
scaled_df[X.columns]=mms.fit_transform(X)
scaled_df['class_0']=y
scaled_df
```

**Model Development**

```python
X=scaled_df.drop("class_0",axis=1)
y=scaled_df['class_0']
```

```python
print(scaled_df.shape)
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print(scaled_df.shape)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

**Linear Regression**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

**Model_prediciton**``

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
import warnings
warnings.filterwarnings('ignore')
lr.predict([X_test.values[0],X_test.values[1]])
```

```python
y_pred
```

```python
new_df=pd.DataFrame(X_test)
new_df['y_actual']=y_test
new_df['y_predicted']=y_pred
new_df['Error']=y_test-y_pred
new_df['square Error']=np.square(y_test-y_pred)
total_error=np.sum(np.square(y_test-y_pred))
total_error
```

```python
mean_squared_error=total_error/len(X_test)
mean_squared_error
```

**Mean Squared Error Method**

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```

```python
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
rmse
```
